## Quick Analysis with charts / Setting up df for future use

```python
#cfs_stats= cfs_stats.drop(columns=['TransactionID'])

cfs_copy = cfs.copy()

cfs_copy = cfs_copy.drop(columns=['TransactionID'])
column_sums = {}

for column in cfs_copy.columns:
    column_name = column
    column_sum = cfs[column_name].sum()
    column_sums[column] = column_sum


print("Column_sums: ")

for column, total in column_sums.items():
    print(f"{column}: {total}")

most_purchased_item = max(column_sums, key=column_sums.get, )
least_purchased_item = min(column_sums, key = column_sums.get)
print("The most purchased menu item was" , most_purchased_item, "and the least "
      "purchased item was", least_purchased_item, ".")
```

```python
#column_sums_df = pd.DataFrame(list(column_sums.items(total), columns=column_sums.items())
column_sums_df = pd.DataFrame(list(column_sums.items()), columns=['Menu Items', 'Total Quantity Sold'])
column_sums_df = column_sums_df.sort_values(by='Total Quantity Sold', ascending= False )
column_sums_df
```

```python
plt.figure(figsize=(8,6))
bars = plt.barh(column_sums_df['Menu Items'],column_sums_df['Total Quantity Sold'], )
plt.bar_label(bars)
plt.xlabel("Quantity of Sold", fontweight = 'bold')
plt.ylabel("Menu Items", fontweight = 'bold')
plt.show()
```

## Correlations

```python
# Finding a correlations among the features

correlation_matrix = cfs.corr()

print("\nCorrelation Matrix: ")
print(correlation_matrix.head(5))
```

```python
# Heatmap of the Correlation Matrix

plt.figure(figsize = (8, 6))
sns.heatmap(correlation_matrix, annot= True, cmap=
            'coolwarm', linewidths=0.5, annot_kws={"size": 5})
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Find the highest correlation (excluding the diagonal)
correlation_matrix_no_diag = correlation_matrix.where(~np.eye(correlation_matrix.shape[0], dtype=bool))
max_corr = correlation_matrix_no_diag.max().max()

# Find the pair of variables with the highest correlation
max_corr_pair = correlation_matrix_no_diag.stack().idxmax()

print(f"\nHighest correlation is between {max_corr_pair[0]} and {max_corr_pair[1]} with a value of {max_corr}")
```

### Quick prep time and profit calculations per transaction

```python
for index, row in cfs.iterrows():
    for column in cfs.columns:
        if column not in ['TransactionID', 'Prep Time per Transaction']:
            value = row[column]
            if value > 0:
                if column in cshop_ops.columns:
                    prep_time_constant = cshop_ops.loc[3, column]
                    prep_time = value * prep_time_constant
                    cfs.at[index, 'Prep Time per Transaction'] += prep_time
                    if prep_time > 15:
                        cfs.at[index, 'Prep Time per Transaction'] ==15
cfs
```

```python
for index, row in cfs.iterrows():
    for column in cfs.columns:
        if column not in ['TransactionID', 'Prep Time per Transaction','Profit per Transaction']:
            value = row[column]
            if value > 0:
                if column in cshop_ops.columns:
                    profit_constant = cshop_ops.loc[2, column]
                    profit_per_item = value * profit_constant
                    cfs.at[index, 'Profit per Transaction'] += profit_per_item

cfs                
```

## Pareto Analysis (Quantity Sold)

```python
cfs_summary_sold = column_sums_df
cfs_summary_sold
```

```python
cfs_summary_sold['Cumulative Sum'] = cfs_summary_sold['Total Quantity Sold'].cumsum()
cfs_summary_sold['Cumulative Percentage'] = 100 * cfs_summary_sold['Cumulative Sum'] / cfs_summary_sold['Total Quantity Sold'].sum()

pareto_threshold = 80
top_20percent_items_sold = cfs_summary_sold[cfs_summary_sold['Cumulative Percentage']<= pareto_threshold]
top_20percent_items_sold
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(top_20percent_items_sold['Menu Items'], top_20percent_items_sold['Total Quantity Sold'], color='blue')
plt.plot(top_20percent_items_sold['Menu Items'], top_20percent_items_sold['Cumulative Percentage'], color='red', marker='o')
plt.xlabel('Product')
plt.ylabel('Total Quantity Sold')
plt.title('Pareto Chart')
plt.xticks(rotation=90)
plt.show()
```

### Pareto Analysis (Which menu items produce top 80% of sales?)

```python
cfs_summary_sold.loc[:,'Total Profit Per Product'] = 0
cfs_summary_sold

```

```python
for index, row in cfs_summary_sold.iterrows():
    for column in cfs_summary_sold.columns:
        if column not in ['Cumulative Sum', 'Cumulative Percentage','Total Profit Per Product']:
            menu_item = row['Menu Items']
            quantity_sold = row['Total Quantity Sold']
            for column in cshop_ops.columns:
                if column == menu_item:
                    profit_constant = cshop_ops.loc[2, menu_item]
                    cfs_summary_sold.at[index, 'Total Profit Per Product'] = profit_constant * quantity_sold
            

cfs_summary_sold  
```

```python
cfs_summary_sold =cfs_summary_sold.sort_values(by='Total Profit Per Product', ascending= False )
cfs_summary_sold
```

```python
cfs_summary_sold['Cumulative Sum'] = cfs_summary_sold['Total Profit Per Product'].cumsum()
cfs_summary_sold['Cumulative Percentage'] = 100 * cfs_summary_sold['Cumulative Sum'] / cfs_summary_sold['Total Profit Per Product'].sum()

pareto_threshold = 22
top_20percent_profit_from_items = cfs_summary_sold[cfs_summary_sold['Cumulative Percentage']<= pareto_threshold]
top_20percent_profit_from_items
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(cfs_summary_sold['Menu Items'], cfs_summary_sold['Total Profit Per Product'], color='blue')
plt.plot(cfs_summary_sold['Menu Items'], cfs_summary_sold['Cumulative Percentage'], color='red', marker='o')
plt.xlabel('Product')
plt.ylabel('Total Quantity Sold')
plt.title('Pareto Chart')
plt.xticks(rotation=90)
for bar in bars[-14:]:
    bar.set_alpha(0.3)
plt.show()
```
